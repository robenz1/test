```python
aggregated_data_smenah = new_data.groupby('ФИО сотрудника')['smena/h'].mean().reset_index()

sorted_data_smenah = aggregated_data_smenah.sort_values('smena/h', ascending=False)

top_10_smenah_head = sorted_data_smenah.head(10)

top_10_smenah_tail = sorted_data_smenah.tail(10)

def add_annotations(ax):
    for p in ax.patches:
        ax.annotate(f"{p.get_width():.1f}",
                    (p.get_width(), p.get_y() + p.get_height() / 2),
                    xytext=(5, 0),
                    textcoords='offset points',
                    ha='left', va='center')

plt.figure(figsize=(10, 6))
ax_head = sns.barplot(x='smena/h', y='ФИО сотрудника', data=top_10_smenah_head)
add_annotations(ax_head)
plt.title('Топ-10 сотрудников с наибольшим количеством часов в смену')
plt.xlabel('Отношение часов к сменам')
plt.ylabel('ФИО сотрудника')
plt.show()

plt.figure(figsize=(10, 6))
ax_tail = sns.barplot(x='smena/h', y='ФИО сотрудника', data=top_10_smenah_tail)
add_annotations(ax_tail)
plt.title('Топ-10 сотрудников с наименьшим количеством часов в смену')
plt.xlabel('Отношение часов к сменам')
plt.ylabel('ФИО сотрудника')
plt.show()
```

#Анализ складов

1. Показатель рабочих часов на сотрудника по складу

```python
total_hours_per_warehouse = new_data.groupby('Транзитный склад').agg({
    'Кол-во отработанных часов по сотруднику': 'sum',
    'ФИО сотрудника': 'nunique'
}).rename(columns={'ФИО сотрудника': 'Количество сотрудников'})

total_hours_per_warehouse['Часов на человека в месяц'] = round(total_hours_per_warehouse['Кол-во отработанных часов по сотруднику'] / total_hours_per_warehouse['Количество сотрудников'],2)

#median_hours_per_person = total_hours_per_warehouse['Часов на человека'].median()
total_hours_per_warehouse = total_hours_per_warehouse.sort_values('Часов на человека в месяц', ascending=False)
total_hours_per_warehouse
```

```python
plt.figure(figsize=(14, 8))
bar=sns.barplot(x=total_hours_per_warehouse['Часов на человека в месяц'], y=total_hours_per_warehouse.index, palette='viridis')
add_annotations(bar)
plt.title('Часов на человека в месяц по складам')
plt.xlabel('Часов на человека в месяц')
plt.ylabel('Транзитный склад')
plt.show()
```

Эту метрику мы можем рассматривать, принимая в условие то, что идет аналитика объемов поставок и нагруженности на штат. Да смены у сотрудников не нормированы, и люди работают разное время в месяц. Эта метрика нам нужна для регуляции размера штата, к примеру склады в Москвы с штатом сотрудников 40+ имеют показатель рабочих часов на сотрудника ненамного выше, чем склад в Ростове-на-Дону, где штат всего 11. Так же следует обратить внимание на склады в Челябинске и Туле, где штат сотрудников одинаковый по размеру, но показатель сильно разнится. Следует варьировать размер штата сотрудников на складе или улучшать рабочий процесс, для более точных рекомендаций необходимы дополнительные данные , как минимум, по временным отрезкам и нормам выполнения рабочего процесса.

3. Количество отсканированных паков в час на складе

```python
pack_per_hour = new_data.groupby('Транзитный склад').apply(
    lambda x: x['Общее кол-во паков обработанных на всем складе'].mean() / x['Кол-во отработанных часов по сотруднику'].sum()
).reset_index(name='Паков в час')
pack_per_hour['Паков в час'] = pack_per_hour['Паков в час'].apply(ceil)
pack_per_hour = pack_per_hour.sort_values('Паков в час',ascending=False)
pack_per_hour
```

```python
plt.figure(figsize=(13, 8))
bar = sns.barplot(x='Паков в час', y='Транзитный склад', data=pack_per_hour, palette='viridis')
add_annotations(bar)
plt.title('Количество паков в час на складе')
plt.xlabel('Паков в час')
plt.ylabel('Транзитный склад')
plt.show()
```
